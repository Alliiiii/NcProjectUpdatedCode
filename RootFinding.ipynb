# Root Finding Method With Comparision Of Graph

```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd

RF=0;
NR=0;
SC=0;
BM=0;

def func( x ): 
    return ((x**2)+5*x-3); 
 

def derivFunc( x ):
    h=(derivative(func,x))
    return h

def f(x): 
    f = func(x) 
    return f; 
  
  
def newtonRaphson( x ,E, n):
    global NR
    newtonCount=0;
    h = func(x) / derivFunc(x) 
    while abs(h) >= E: 
        
        if(derivFunc(x)==0):
            print("Not Possible")
            break;
        else:
            h = func(x)/derivFunc(x) 
        
            x = x - h 
     
    
            newtonCount=newtonCount+1;
            
        if(newtonCount>n):
            break
        
    
    NR=newtonCount
    if(newtonCount>n):
        print("Could Not Found Solution even in ",n, "iterations")
    else:
        print("The value of the root By Newton rephson method is : ", "%.4f"% x) 
        print("Counts : ",newtonCount)
    
    

    
def bisection(a,b,E, n):
    global BM
    bisectionCount=0;
    if (func(a) * func(b) >= 0):
        print("Here f(a)>0 and f(b)>0 So, Solution is not possible. Try Again...")
        return
   
    c = a 
    while ((b-a) >= E): 
  
        c = (a+b)/2
   
        if (func(c) == 0.0): 
            break
   
        if (func(c)*func(a) < 0): 
            b = c 
        else: 
            a = c 
        bisectionCount=bisectionCount+1
        
        if(bisectionCount>n):
            break;
    
    BM=bisectionCount
    
    if(bisectionCount>n):
        print("Could Not Found Solution even in ",n, "iterations")
    else:
        print("The value of root is : ","%.4f"%c) 
        print("Counts : ",bisectionCount)
    

    
    
    

def secant(x1, x2, E, nn): 
    global SC
    n = 0; xm = 0; x0 = 0; c = 0;  
    
    if(f(x1)*f(x2)<0):
        while True:  

            x0 = ((x1 * f(x2) - x2 * f(x1)) / 
                            (f(x2) - f(x1)));  

            c = f(x1) * f(x0);  

            x1 = x2;  
            x2 = x0;  

            n += 1; 
            
            if(n>nn):
                break;

            if (c == 0):  
                break;  
            xm = ((x1 * f(x2) - x2 * f(x1)) / 
                            (f(x2) - f(x1))); 

            if(abs(xm - x0) < E): 
                break; 

        SC=n
        
        if(n>nn):
            print("Could Not Found Solution even in ",nn, "iterations")
        else:
            print("Root of the given equation =",  
                                   round(x0, 4));  
            print("No. of iterations = ", n);  
    
    else:
        print("Here f(a)>0 and f(b)>0 So, Solution is not possible. Try Again...")
  
        
        
        

  
    
    
    
  


def regulaFalsi(a,b,TOL,N):
    global RF
    i = 1
    FA = f(a)
     
    if(f(a)*f(b)<0):

        while True:
            p = (a*f(b)-b*f(a))/(f(b) - f(a))
            FP = f(p)

            if(FP == 0 or np.abs(f(p)) < TOL):
                break



            i = i + 1

            if(FA*FP > 0):
                a = p
            else:
                b = p
            
            if(i>N):
                break
        
        if(i>N):
            print("Could Not Found Solution even in ",nn, "iterations")
        else:
            print("The value of root is : " , '%.4f'%a)
            print("Count : ",i)
        RF=i;
        
       
        
    else:
        print("Here f(a)>0 and f(b)>0 So, Solution is not possible. Try Again...")
    return

def CallMethods(a,b,E,N):
    print("----------Bisection-------------")
    bisection(a, b,E,N)  
 
    print("-----------Regula Falsi------------")
    regulaFalsi(a,b,E,N);

    print("---------Newton Raphson--------------")
    newtonRaphson(a,E,N)

    print("------------Secant-----------")
    secant(a, b, E,N);


 
    
    return


def draw_Graph():
    
    MethodName = ['Bisection Method','Regula Falsi','Newton Rephson','Secant Method']
   
    Count = [BM,RF,NR, SC]
     
    plt.plot(MethodName, Count, color='g')
    plt.xlabel('Method Name')
    plt.ylabel('No. of iteration')
    plt.title('Comparision of root finding Methods')
    plt.show()
    return


def draw_Bar_Graph():
    
    s = pd.Series(
        [BM,RF,NR, SC],
        index = ['Bisection Method','Regula Falsi','Newton Rephson','Secant Method']
        )


    plt.title("Comparision of Root Finding Methods")
    plt.ylabel('No of Iteration')
    plt.xlabel('Mathod Name')


    ax = plt.gca()
    ax.tick_params(axis='x', colors='blue')
    ax.tick_params(axis='y', colors='red')
   
    my_colors = 'Black'  

    s.plot( 
        kind='bar',
        color=my_colors,

    )

    plt.show()
    return



a =int(input("Please enter value of x \n"))
b =int(input("Please enter value of x1 \n"))
E =float(input("Please enter value of Tolerence \n"))
n=int(input("Please enter value of iteration \n"))


CallMethods(a,b,E,n)

draw_Bar_Graph()
draw_Graph()
```
